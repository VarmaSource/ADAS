```python
import cv2
import numpy as np

# Load YOLO model and weights
net = cv2.dnn.readNet("yolov3.weights", "yolov3.cfg")
classes = []
with open("coco.names", "r") as f:
    classes = [line.strip() for line in f.readlines()]
layer_names = net.getLayerNames()
output_layers = [layer_names[i[0] - 1] for i in net.getUnconnectedOutLayers()]

# Load video
cap = cv2.VideoCapture("D:\VN20240129_174924.mp4")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    height, width, channels = frame.shape

    # Detect objects
    blob = cv2.dnn.blobFromImage(frame, 0.00392, (416, 416), (0, 0, 0), True, crop=False)
    net.setInput(blob)
    outs = net.forward(output_layers)

    # Process detections
    for out in outs:
        for detection in out:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if confidence > 0.5:  # Adjust confidence threshold as needed
                # Object detected
                center_x = int(detection[0] * width)
                center_y = int(detection[1] * height)
                w = int(detection[2] * width)
                h = int(detection[3] * height)

                # Rectangle coordinates
                x = int(center_x - w / 2)
                y = int(center_y - h / 2)

                # Draw bounding box
                cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
                cv2.putText(frame, classes[class_id], (x, y + 30), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Show frame
    cv2.imshow("Object Detection", frame)

    # Break the loop if 'ESC' key is pressed
    if cv2.waitKey(30) == 27:
        break

# Release the capture and close all windows
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Step 6: Real-time Video Processing
cap = cv2.VideoCapture('D:/VID20230516053330.mp4')

# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output.avi', fourcc, 20.0, (int(cap.get(3)), int(cap.get(4))))

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Preprocess the frame for the model
    img = cv2.resize(frame, (224, 224))
    img = np.expand_dims(img, axis=0)
    img = img / 255.0  # Normalize pixel values

    # Make predictions using the trained model
    predictions = loaded_model.predict(img)

    # Get the predicted class
    predicted_class_index = np.argmax(predictions)
    classes = ['Speed Hump', 'Plain Road', 'Pothole']
    predicted_class_label = classes[predicted_class_index]

    # Visualize the result on the frame
    cv2.putText(frame, f'Prediction: {predicted_class_label}', (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

    # Write the frame with the prediction to the output video
    out.write(frame)

    # Display the result
    cv2.imshow('Object Recognition', frame)

    # Break the loop if 'ESC' key is pressed
    if cv2.waitKey(30) == 27:
        break

# Release the capture and close all windows
cap.release()
out.release()
cv2.destroyAllWindows()
```
